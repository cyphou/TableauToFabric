# Data Transformations — filtering

This notebook materialises calculated columns (row-level
Tableau formulas) as physical columns in the Lakehouse
Delta table, so the Semantic Model can reference them
directly via DirectLake.

**Generated:** 2026-03-05 10:47


```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
from pyspark.sql.types import *
from pyspark.sql.window import Window
from datetime import datetime

print(f"Transformations started: {datetime.now()}")
```

## Calculated Columns (materialised)

1 calculated column(s) will be added
to the `testdata` Delta table.


```python
# Read the main table from Lakehouse
df = spark.read.format("delta").table("testdata")
print(f"Rows: {df.count()}, Columns: {len(df.columns)}")
```

### `SHOW` (string)

**Tableau formula:** `IF [BurstoutSet] THEN "Selected" ELSE "Not Selected" END`

```python
df = df.withColumn("SHOW", F.when(F.col("BurstoutSet"), "Selected").otherwise("Not Selected"))
```

```python
# Overwrite the Delta table with the new columns
df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("testdata")
print(f"✓ Written {len(df.columns)} columns to testdata")
```
