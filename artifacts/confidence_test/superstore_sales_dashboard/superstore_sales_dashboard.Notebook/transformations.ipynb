# Data Transformations — superstore_sales_dashboard

This notebook materialises calculated columns (row-level
Tableau formulas) as physical columns in the Lakehouse
Delta table, so the Semantic Model can reference them
directly via DirectLake.

**Generated:** 2026-03-05 10:47


```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
from pyspark.sql.types import *
from pyspark.sql.window import Window
from datetime import datetime

print(f"Transformations started: {datetime.now()}")
```

## Calculated Columns (materialised)

1 calculated column(s) will be added
to the `extract` Delta table.


```python
# Read the main table from Lakehouse
df = spark.read.format("delta").table("extract")
print(f"Rows: {df.count()}, Columns: {len(df.columns)}")
```

### `Grouping Based on Unit Price` (string)

**Tableau formula:** `IF [Unit Price] <= 2500 then 'A'
ELSEIF [Unit Price] <=5000 then 'B'
ELSE 'C' END`

```python
df = df.withColumn("Grouping Based on Unit Price", F.when(F.col("Unit Price") <= 2500, 'A'
ELSEIF F.col("Unit Price") <=5000 then 'B').otherwise('C'))
```

```python
# Overwrite the Delta table with the new columns
df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("extract")
print(f"✓ Written {len(df.columns)} columns to extract")
```
