```python
# ── Setup ──────────────────────────────────────────────
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
from datetime import datetime

# Fabric Lakehouse is automatically mounted
LAKEHOUSE_NAME = "global_superstores_db_Lakehouse"
print(f"ETL Pipeline started: {datetime.now()}")
```

## Data Ingestion

Loading 1 table(s) from source systems.


### Table: `Orders` (Excel)

```python
# Read Excel file (requires com.crealytics:spark-excel)
df_orders = spark.read \
    .format("com.crealytics.spark.excel") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("dataAddress", "\'Orders\'!A1") \
    .load("Files//Users/bs/Mac Drive/MSBA/INFO584_01SP21/Datasets/Global Superstore.xlsx")

# Preview
df_orders.printSchema()
print(f"Rows: {df_orders.count()}")

# Write to Lakehouse Delta table
df_orders.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("orders")
print(f"✓ Written to Lakehouse: orders")
```
