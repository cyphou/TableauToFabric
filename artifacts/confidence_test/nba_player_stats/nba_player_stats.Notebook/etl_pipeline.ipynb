```python
# ── Setup ──────────────────────────────────────────────
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
from datetime import datetime

# Fabric Lakehouse is automatically mounted
LAKEHOUSE_NAME = "nba_player_stats_Lakehouse"
print(f"ETL Pipeline started: {datetime.now()}")
```

## Data Ingestion

Loading 1 table(s) from source systems.


### Table: `Sheet1` (Excel)

```python
# Read Excel file (requires com.crealytics:spark-excel)
df_sheet1 = spark.read \
    .format("com.crealytics.spark.excel") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("dataAddress", "\'Sheet1\'!A1") \
    .load("Files/Data/Datasets/nba_players.xlsx")

# Preview
df_sheet1.printSchema()
print(f"Rows: {df_sheet1.count()}")

# Write to Lakehouse Delta table
df_sheet1.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("sheet1")
print(f"✓ Written to Lakehouse: sheet1")
```
